```python
import pandas as pd 
import numpy as np
```

```python
def f(x, k, a, b):
    return 1 - (1 / (1 + np.exp(-k * (x - a)))) + (1 / (1 + np.exp(-k * (x - b))))
```

```python
Y = pd.Series([f(x, 0.5, 20, 40) for x in range(100)])
Y
```

```python
Y.plot()
```
